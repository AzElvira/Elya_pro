### Посчитаем, сколько книг вышло после 1 января 2000 года;

```python
query = '''
SELECT 
    COUNT(DISTINCT book_id) AS cnt
FROM
    books
WHERE
    publication_date :: date > '2000-01-01'; 
    '''
pd.io.sql.read_sql(query, con = engine)
```

Учитывая что в таблице 1000 книг-80% из них выпущены в последние 20 лет.

### Для каждой книги посчитаем количество обзоров и среднюю оценку

```python
query = '''
SELECT
    books.book_id,
    COUNT(DISTINCT reviews.review_id) AS "Колличество обзоров",
    AVG(ratings.rating) AS "Средняя оценка"
FROM
    books
RIGHT JOIN ratings ON ratings.book_id = books.book_id
RIGHT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY
    books.book_id
ORDER BY
    "Колличество обзоров" DESC;     
;    
    '''
pd.io.sql.read_sql(query, con = engine) 
```

"Сумерки"-любимое чтиво подростков, имеет больше всех обзоров со стороны читателей, однако не самый высокий балл. Очевидно данное произведение никого не оставило равнодушным.
"Хоббит", "над пропастью во ржи", при практически одинаковом колличестве обзоров находятся примерно в одном рейтинговом поле.
Ну и, конечно, поттертиана-ожидала увидеть рейтинг не ниже 4.5

Все это больше подростковая литература. Увеличенное колличество отзывов молодого поколения говорит о более интенсивном использовании интернета, в том числе путем оставления пользовательских отзывов, но не является доказательством их большего приобщения к литературе. Взрослое поколение очевидно не считает нужным оставлять свои комментарии в сети. Поэтому объективной популярности иной литературы, направленной на читателей старшего возраста, не в полном объеме находит отражение в количестве отзывов.

### Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так мы исключим из анализа брошюры

```python
query = '''
SELECT
    publishers.publisher AS "Наименование Издательства",
    COUNT(books.book_id) AS "Колличество книг"
FROM
    books
INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
WHERE num_pages > '50'
GROUP BY 
    publisher
ORDER BY
    "Колличество книг" DESC; 
    '''
pd.io.sql.read_sql(query, con = engine)
```

### Определим автора с самой высокой средней оценкой книг — учитывая только книги с 50 и более оценками;

```python
query = '''
SELECT 
    authors.author AS "Автор книги",
    Sub."Средняя оценка"
FROM
    authors
RIGHT JOIN (SELECT
                books.author_id,
                AVG(ratings.avg) as "Средняя оценка"
            FROM 
                books
            RIGHT JOIN ( SELECT
                            book_id,
                            COUNT(rating) as cnt,
                            AVG(rating) as avg
                        FROM
                            ratings
                        GROUP BY
                            book_id
                        HAVING COUNT(rating)>=50) as ratings 
           ON ratings.book_id= books.book_id
           GROUP BY
               books.author_id) AS Sub
ON Sub.author_id=authors.author_id
ORDER BY
    Sub."Средняя оценка" DESC
    LIMIT 5; 
    '''
pd.io.sql.read_sql(query, con = engine)
```

### Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''
SELECT 
    AVG(Asq) AS "Среднее колличество обзоров"
FROM
    (SELECT 
           (SELECT 
                COUNT(*)
                FROM reviews
                WHERE
                    username=ratings.username) as Asq
    FROM
        ratings

    GROUP BY 
        username
    HAVING 
        COUNT(*)>'50') as Br;  
 
    '''
pd.io.sql.read_sql(query, con = engine)
```

Исходя из такого среднего колличества обзоров можно делать выводы о покупке данного продукта. Вообще, чем больше отзывов и оценок тем лучше обратная связь и помощь для "новичков"(родители подростков подбирающих литературу), так и для опытных пользователей.
